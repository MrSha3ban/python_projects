```python
#ch  انشاء دالة تسجيل الاختيار في  المتغير 


def choice (lst_lst,i):
    global ch
    ch=input(f"\n \n برجاء الاختيار من القائمة : \n 1. {lst_lst[i][0]} \n 2. {lst_lst[i][1]} \n")
    while ch  not in ["1","2"]:
       ch=input(f"\n \n برجاء الاختيار بشكل صحيح : \n 1. {lst_lst[0][0]} \n 2. {lst_lst[0][1]} \n")
    else:
        ch=int(ch)-1
        print("لقد اخترت الاختيار",ch+1)

#إنشاء دالة تقوم بإضافة نتيجة السؤال الى النص  الذي يتجمع فيه حروف  نتائج الاختبار


def score(lst,cat):
    global txt
    txt=""
    for i in lst:
        choice(lst,lst.index(i))
        txt+=cat[lst.index(i)][ch]
    
score(lsta,cat)
```

```python
#تجميع النسب الخاصة بكل نوع من نتيجة الاختبار

result={"I":txt.count("I"),"E":txt.count("E"),"N":txt.count("N"),"S":txt.count("S"),"F":txt.count("F"),
"T":txt.count("T"),"P":txt.count("P"),"J":txt.count("J")}

test_result=""


# المفاضلة بين كل زوج من الانواع المتضادة بناء على النسب


for x in result.keys():
    if (list(result.keys()).index(x)+1) %2 ==0:
        if result.get(x) > result.get(list(result.keys())[list(result.keys()).index(x)-1]):
            test_result+=x
        else:
            test_result+=list(result.keys())[list(result.keys()).index(x)-1]
            
            
    else:
        continue



#طباعة النتيجة في شكل مناسب
clear_output()
display(Markdown(f"<span style='font-size:48px; font-weight:bold'>{test_result} : نتيجة الاختبار الخاصة بك هي</span>"))   
```
